```python
import numpy as np
from scipy.stats import linregress

plt.rcParams['lines.linewidth'] = 1
plt.rcParams['scatter.marker'] = "x"
plt.rcParams['font.size'] = 12

errors = []
for _ in range(1):
    x = np.array([0, 1, 2, 3, 4, 5], dtype=np.float64)
    y = np.array([50, 50, 50, 50, 50, 50], dtype=np.float64)
    noise = np.random.triangular(-1, 0, 1, size=len(x))
    y += noise
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    errors.append(std_err)
mean_error = np.mean(errors)
print("Mean Error:", mean_error)

import matplotlib.pyplot as plt
plt.figure(figsize=(4,3))
plt.ylim(47, 51.2)
#plt.errorbar(x, y, yerr=1, fmt='o', label='Messung mit\n Rauschen und Fehlerbalken')

plt.errorbar(x, np.array([50, 50, 50, 50, 50, 50]), yerr=np.ones_like(x) * 0.4, fmt="x", label='Messwerte mit Fehlerbalken')
plt.scatter(x, np.array([50, 50, 50, 50, 50, 50]) + noise, color='red', label='Messwerte mit Rauschen')
plt.plot(x, slope * x + intercept, label='linearer fit')
plt.xlabel('$t$ in min')
plt.ylabel('$V$ in ml')
plt.legend()
plt.show()
```
